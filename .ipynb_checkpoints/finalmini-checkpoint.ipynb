```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from os import listdir
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Input,ZeroPadding2D,Conv2D,BatchNormalization,Activation,MaxPooling2D,Flatten,Dense
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import TensorBoard,ModelCheckpoint
import time
```

function for cropping image

```python
def cropbrain(img):

     #conv img to grey 
    grey_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)    

     #blur the grey img 
    blur_img=cv2.GaussianBlur(grey_img,(5,5),0)

     #create binary image
    thresh=cv2.threshold(blur_img,45,255,cv2.THRESH_BINARY)[1]
    
    #remove noise and find gaps in contour
    erode_img=cv2.erode(thresh,None,iterations=2)
    dil_img=cv2.dilate(erode_img,None,iterations=2)

    #find contour in dilate img
    contour=cv2.findContours(dil_img.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)[0]

    #find contour with maximum area
    large_contour=max(contour,key=cv2.contourArea)

    #find extereme points
    left_point=tuple(large_contour[large_contour[:,:,0].argmin()][0])
    right_point=tuple(large_contour[large_contour[:,:,0].argmax()][0])
    top_point=tuple(large_contour[large_contour[:,:,1].argmin()][0])
    bot_point=tuple(large_contour[large_contour[:,:,1].argmax()][0])

    newimg=img[top_point[1]:bot_point[1],left_point[0]:right_point[0]]

    return newimg
    
```

load all images in a directory

```python
def load(dirli,imgh,imgw):
    x=[]
    y=[]
    for label,li in enumerate(dirli):
        for file in listdir(li):
            img=cropbrain(cv2.imread(li+'/'+file))
            img=cv2.resize(img,dsize=(imgh,imgw),interpolation=cv2.INTER_CUBIC)

            #normalising the img ranging from 0-255
            x.append(img/255)
            y.append(label)

    #shuffling the data using sklearn library for good distribution in validation and training set
    x,y=shuffle(np.array(x),np.array(y),random_state=30)
    return x,y
```

plotting the image to check the working of crop function

```python
eg=cv2.imread('yes/p (1).jpg')
plt.subplot(1,2,1)
plt.title('original')
plt.imshow(eg)                 
eg2=cropbrain(eg)
plt.subplot(1,2,2)
plt.title('cropped')
plt.imshow(eg2);                                                             #;to remove <matplotlib.image.AxesImage at 0x...> above img
```

plotting some examples for debugging and visual inspection

```python
def plotimg(x,y,n=20):
    for label in [0,1]:
        img=x[y==label][:n]
        col_n=10
        row_n=int(n/col_n)
        plt.figure(figsize=(20,5))
        for i,imge in enumerate(img,1):
            plt.subplot(row_n, col_n, i)
            plt.imshow(imge)
            plt.axis('off')
            plt.suptitle(f"Brain Tumor: {'Yes' if label == 1 else 'No'}")
        plt.show()
```

splitting the data into training ,testing and validation set

```python
def spltdata(x, y, tsize=0.2):
    xtrain,xtemp,ytrain,ytemp=train_test_split(x,y,train_size=tsize,random_state=42)    #train= 60%
    xval,xtest,yval,ytest=train_test_split(xtemp,ytemp,train_size=0.5,random_state=42)  #test / val =20%
    return xtrain,ytrain,xval,yval,xtest,ytest
```

building the model using CNN

```python
def bmodel(imgshape):
    xinput=Input(imgshape)
    x=ZeroPadding2D((2,2))(xinput)
    x=Conv2D(32,(7,7),strides=(1,1),name='conv0',activation='relu')(x)
    x=BatchNormalization(axis=3,name='bn0')(x)                             #helps in accelerating training by stabilizing
    x=Activation('sigmoid')(x)
    x=MaxPooling2D((4,4),name='max0')(x)                                    #reduce spatial dimension
    x=MaxPooling2D((4,4),name='max1')(x)
    x=Flatten()(x)                                                        #conv matrix to vector for fully connection
    
    x=Dense(1,activation='sigmoid',name='fc')(x)                          #adds fully connected layer with neuron and gives biary output
                                                                          #helps in classification

    model=Model(inputs=xinput,outputs=x,name='braintumordetection')
    return model
```

train the model using tensorflow

```python
def trainmodel(model,xtrain,ytrain,xval,yval,n=12):
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')     #it compiles the model and accuracy as a metric to monitor during training
    logfile=f'braintumordetection_cnn{int(time.time())}'
    tensorboard=TensorBoard(log_dir=f'logs/{logfile}')  #to monitor the training process and store in log file
    filepath='models/cnn-parameters-improvement-{epoch:02d}-{val_accuracy:.2f}'
    checkpoint=ModelCheckpoint(filepath,monitor='val_accuracy',verbose=1,save_best_only=True,mode='max')
    history=model.fit(xtrain,ytrain,batch_size=32,epochs=n,validation_data=(xval,yval),callbacks=[tensorboard,checkpoint])
    return history
```

now lets begin

load and preprocess the data

```python
imgh,imgw=(240,240)
dirli=['yes','no']
x,y=load(dirli,imgh,imgw)
```

visualizing samples

```python
plotimg(x,y)
```

split data

```python
xtrain,ytrain,xval,yval,xtest,ytest=spltdata(x,y,0.2)
```

build and train the model

```python
imgshape=(imgw,imgh,3)
model=bmodel(imgshape)
history=trainmodel(model,xtrain,ytrain,xval,yval,n=12)
model.save('brain_tumor_detection_model.h5')                                     #save the model
```

```python
def plotgraph(history):
    plt.figure(figsize=(12,4))

    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'],label='Train')
    plt.plot(history.history['val_accuracy'],label='Validation')
    plt.title('model accuracy')
    plt.xlabel,plt.ylabel='epochs','accuracy'
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(history.history['loss'],label='Train')
    plt.plot(history.history['val_loss'],label='Validation')
    plt.title('model loss')
    plt.xlabel,plt.ylabel='epochs','loss'
    plt.legend()

    plt.tight_layout()
    plt.show()

plotgraph(history)
```
